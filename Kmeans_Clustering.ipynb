```python
import pandas as pd
```

```python
#small dataset for demo purposes 
data={ 'CustomerID':[1,2,3,4,5,6,7,8],
       'AnnualIncome':[15,16,17,28,30,45,55,60],
       'SpendingScore':[39,81,6,77,40,50,42,30]
     }
```

```python
df = pd.DataFrame(data)
df
```

```python
#import the lib
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#select features for the clustering
X = df[['AnnualIncome','SpendingScore']]
```

```python
X=df[['AnnualIncome','SpendingScore']]  
plt.figure(figsize=(6,4))
plt.scatter(X['AnnualIncome'],X['SpendingScore'],s=100, c='violet', edgecolors='k')
plt.title('Customer Didtribution')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.grid(True)
plt.show()
```

```python
# Apply K-Means Clustering
kmeans = KMeans(n_clusters=3, random_state=0)
df['Cluster'] = kmeans.fit_predict(X)

# Clustered Data Plot
plt.figure(figsize=(6,4))
sns.scatterplot(data=df, x='AnnualIncome', y='SpendingScore', hue='Cluster', palette='Set1', s=100)

# Plotting Centroids
centroids = kmeans.cluster_centers_
plt.scatter(centroids[:, 0], centroids[:, 1], s=200, c='black', marker='X', label='Centroids')

plt.title('Customer Segments (K-Means)')
plt.legend()
plt.grid(True)
plt.show()
```
